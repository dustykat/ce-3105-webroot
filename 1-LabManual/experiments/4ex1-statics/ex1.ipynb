# Structured Inquiry vs. "Cook-and-Look": Two Paths Through the Laboratory

## Overview

Not all experiments are created equal. In engineering education, two styles of experimentation commonly emerge:

- **Scientific-method-directed experiments**, where theory guides the inquiry, predictions are made, and results are critically evaluated to test a hypothesis or validate a model.
- **"Cook-and-look" experiments**, where a procedure is followed step-by-step with little theoretical framing or expectation, often leading to passive observation rather than active discovery.

Understanding the distinction is essential not only for academic success, but for cultivating the mindset of a practicing engineer.

---

## Scientific-Method-Directed Experimentation

### Characteristics:
- Begins with a hypothesis or model derived from theory
- Predictions are made in advance
- Experiments are **designed** to test, calibrate, or refute the model
- Measurements are compared to expectations with quantitative rigor
- Discrepancies lead to refinement of models and deeper understanding

### Notable Successes

- **Galileoâ€™s inclined plane experiments (1600s)**  
  Tested mathematical relationships between acceleration, time, and distance, providing early evidence for the laws of motion.

- **Cavendishâ€™s measurement of the gravitational constant (1797)**  
  Designed to quantify Newtonâ€™s law of universal gravitationâ€”not merely observe attraction.

- **Prandtlâ€™s boundary layer experiments (early 1900s)**  
  Theoretical analysis led to the design of experiments that revealed the structure of viscous flow, revolutionizing fluid mechanics.

### Notable Failures 

- **Michelsonâ€“Morley experiment (1887)**  
  Failed to detect the "aether wind" predicted by classical physics. Though it produced a â€œnull result,â€ it motivated a shift in understanding that led to Einsteinâ€™s theory of relativity.

---

## "Cook-and-Look" Experimentation

### Characteristics:
- Follows a fixed set of steps like a recipe, often from a lab manual
- No hypothesis or model-driven prediction is made
- Data is collected without clear expectations
- Emphasis is on procedure compliance, not interpretation
- Often ends with, â€œThe results didnâ€™t match, but we followed the steps.â€

### Occasional Successes

- **Alexander Flemingâ€™s discovery of penicillin (1928)**  
  Accidental contamination of a petri dish revealed antibacterial effectsâ€”not predicted by theory, but recognized through observation.

- **Discovery of cosmic microwave background (Penzias and Wilson, 1965)**  
  Researchers were troubleshooting antenna noise and stumbled onto the remnant of the Big Bangâ€”without initially understanding the implications.

These examples show that "cook-and-look" can occasionally lead to breakthroughs, but only if the observer is prepared to recognize the unexpected and reframe the context.

### Common Failures

- Countless teaching, research, and industrial laboratories where results are recorded without interpretation.
- Industrial settings where procedures are followed, but underlying phenomena are misunderstood, leading to poor troubleshooting or design decisions

---

## Why Scientific Method Matters in Engineering

Engineers do not merely measureâ€”they model, predict, and verify. Experiments are not recipes; they are **tools to challenge assumptions**, test applicability, and build confidence in our understanding of physical systems.

In this fluids laboratory, we reject the "cook-and-look" mindset. Instead, we adopt a **model-driven approach** that values prediction, observation, comparison, and critical thinking. Your role is not to confirm the manualâ€”it is to test nature and decide whether our models are worthy of your trust.

# Example 1: Flow Measurement with a Venturi

## ðŸ”¬ Part I â€” Mini-Model Statement (Pre-Lab)

### 1. Governing Model or Principle

This experiment is based on the **Bernoulli equation**, which states that for an incompressible, steady, inviscid flow along a streamline, the total mechanical energy is conserved. The pressure drop between the wide and narrow sections of a Venturi meter can be related to the flow velocity.

---

### 2. Model Equation(s)

$\Delta p = \frac{1}{2} \rho \left( V_2^2 - V_1^2 \right) $

Where:  
- $\Delta p$ = pressure difference between upstream and throat  
- $\rho$ = fluid density  
- $V_1, V_2$ = velocities at inlet and throat (related via continuity)

From continuity: $Q = A_1 V_1 = A_2 V_2$

---

### 3. Assumptions

- Steady flow
- Incompressible fluid (water)
- No energy loss due to friction or turbulence
- Accurate pressure measurements via manometer

---

### 4. Prediction of Behavior

We expect the pressure at the throat to be lower than at the inlet due to increased velocity. Using measured pressure drop and known cross-sectional areas, we can calculate discharge $Q$ and compare it to direct volumetric measurements.

---

### 5. Evaluation Criteria

We will compare the discharge calculated from the Bernoulli-based model to the flow rate measured by time-to-fill a graduated tank. Agreement within Â±5% will be considered acceptable.

---

## Part II â€” Post-Lab Reflection

### 1. Agreement with the Model (5 pts)

The model-based flow rate was within 4.2% of the directly measured volumetric flow rate for all trials. A comparison plot (Figure 3) shows a strong correlation (RÂ² = 0.993) between predicted and measured values.

---

### 2. Explanation of Discrepancies (5 pts)

Minor discrepancies likely arose from:
- Inexact manometer readings due to parallax
- Slight turbulence or energy loss not accounted for in the model
- Assumption of ideal flow conditions may not be fully valid in the experimental apparatus

---

### 3. Implications and Next Steps (5 pts)

This experiment confirms that the Bernoulli-based model provides a valid estimate of flow rate in a Venturi meter under controlled conditions. Future experiments could investigate how accuracy degrades with increasing turbulence or cavitation, or apply a discharge coefficient $ C_d$ to improve model fit.
