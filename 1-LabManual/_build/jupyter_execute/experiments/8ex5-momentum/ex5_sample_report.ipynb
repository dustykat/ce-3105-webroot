## Data Reduction

The table below lists important physical dimensions in the experiment(s)

|Item|Value|Units|
|----:|----:|----:|
|$\rho$|$1000$|$\frac{kg}{mm^3}$|
|$g$|$9.81$|$\frac{m}{s^2}$|
|jet diameter|$10$|$mm$| 
|nozzle-to-vane distance|$35$|$mm$|              
|Jockey mass|$600$|$mg$|
|Vane-to-pivot distance|$150$|$mm$|

The generic reduction procedure is to:

1. Convert flow rate to mass flow rate, $\dot m~\frac{kg}{s}$, from $\dot m = \rho ~ Q$
2. Calculate the jet velocity, $u_j~\frac{m}{s}$, from $u = \frac{4Q}{\pi d^2}$ where $d$ is the nozzle diameter.
3. Calculate the deflected velocity, $u_0~\frac{m}{s}$ from $u_0 = \sqrt{u_j^2 - 2gs}$ where $s$ is the nozzle-to-vane distance.  This velocity accounts for the conversion of kinetic energy to gravitational potential as the jet rises the distance $s$ before impact with the vane.
4. Calculate rate of delivery of momentum, $\dot m u_0 ~ \frac{kg~m}{s}$
5. Calculate force, $F~\frac{kg~m}{s^2}$.  The values are obtained from the data table(s) $F = \frac{M \cdot g \cdot (a + y)}{a}$, where $y$ is the jockey weight distance from the repeatable zero, $a$ is the pivot to vane attachment location, and $M$ is the mass of the jockey weight.
6. Create a plot of force, $F$ (y-axis) vs. rate of delivery of momentum, $\dot m u_0$(x-axis)
  - Create a trendline for the data points
  - The slope of the trendline indicates the coefficient for theoretical force

## Results and Discussion

For each type of vane, the expected force can be estimated using the momentum equation. When we plot force versus momentum and fit a straight line, the slope of that line represents ($1-cos(\beta)$), where $\beta$ iis the flow deflection angle, as shown in the sketch below:

![](sketch.png)

**Flat Plate**

The plot of Force vs. Momentum for the flat plate is shown below with a fitted line in red. The script used to create the graph and fit the line is included in the appendix.

![](plate_plot.png)

The slope of the line is 0.98. This corresponds to a deflection angle of $\frac{\pi}{2.03}$, which is very close to the ideal value of $\frac{\pi}{2}$.

**Hemispherical Cup**

The Force vs. Momentum data for the hemispherical cup is shown in the next plot, again with a red fitted line. The code used to generate the figure is also in the appendix.

![](cup_plot.png)

For this case, the slope is 1.94. That corresponds to a deflection angle of $\frac{\pi}{1.12}$, which is close to the theoretical value of $\frac{\pi}{1}$.

## Sources of Error

There were several sources of error during the experiment:

- Flow measurement issues: At low flow rates, the H1F flow meter sometimes showed zero even when flow was visible. The PRi mass flow meter was able to measure those flows. In most runs, the two meters agreed within about 2%, with the PRi reading slightly lower. A third method was not available to verify the actual flow rate.
- Reading the jockey weight: There could be small errors when reading the mass position or making sure the system returned to the same zero point each time. This probably didnâ€™t matter much at higher flows.
- Flow instability: The flow was observed to fluctuate as the pump delivered pressure, which could add uncertainty to the force and flow readings.

## Conclusion
Overall, the experimental values matched the theoretical predictions closely. The method used in this lab could also be applied to vane shapes where the theoretical force is harder to calculate. The results show that it is possible to determine an effective deflection angle from measurements and use that information for design.

## Appendix-I Python Analysis Scripts

```python
def mass_flow(rho,Q):
    '''function to return mass flow in kg/s
         rho == density kg/m^3
         Q == flow rate m^3/s
         user must make unit conversions in parameter list
    '''
    mass_flow = rho*Q  
    return(mass_flow)

def appliedF(mass,g,a,y):
    '''function to return applied force in Newtons
    '''
    appliedF=mass*g*(a+y)/a
    return(appliedF)
    
import math
# Data from tables
y_plate=[102,86,68,42,19]
q_plate=[35.5,28.4,27.2,22.6,17.3]
y_cup  =[188,140,90,50,20]
q_cup  =[33.1,29.8,24.3,18.2,12.6]

# Constants
rho      = 1000     #kg/m^3
g        = 9.81     #m/s^2
diameter = 0.010    #m  nozzle diameter
s        = 0.035    #m distance between nozzle-vane
Mass     = 600/1000 #kg the jockey mass
a        = 0.150    # m vane-pivot distance

# Calculations
# - allocate memory locations
mdot_plate = [] # mass flow plate
mdot_cup = []   # mass flow cup
vj_plate=[]     # jet velocity plate
vj_cup=[]       # jet velocity cup
vd_plate=[]     # deflection velocity plate
vd_cup=[]       # deflection velocity cup
mom_plate = []  # momentum entering at plate
mom_cup = []    # momentum entering at cup
f_plate = []
f_cup = []
# - populate vectors 
for i in range(len(y_plate)):
    mdot_plate.append(mass_flow(1000,q_plate[i]/1000/60))
    mdot_cup.append(mass_flow(1000,q_cup[i]/1000/60))
    vj_plate.append(4*(q_plate[i]/1000/60)/(math.pi*diameter**2))
    vj_cup.append(4*(q_cup[i]/1000/60)/(math.pi*diameter**2))
    vd_plate.append(math.sqrt(vj_plate[i]**2 - 2*g*s))
    vd_cup.append(math.sqrt(vj_cup[i]**2 - 2*g*s))
    mom_plate.append(mdot_plate[i]*vd_plate[i])
    mom_cup.append(mdot_cup[i]*vd_cup[i])
    f_plate.append(appliedF(Mass,g,a,y_plate[i]/1000))
    f_cup.append(appliedF(Mass,g,a,y_cup[i]/1000))
# - print results for tabulation
print("plate")
for i in range(len(y_plate)):
    print("|",i+1,"|",q_plate[i],"|",round(mdot_plate[i],3),"|",y_plate[i],"|"\
         ,round(vj_plate[i],3),"|",round(vd_plate[i],3),"|",round(mom_plate[i],3),"|",round(f_plate[i],3),"|")
print("cup")
for i in range(len(y_cup)):
    print("|",i+1,"|",q_cup[i],"|",round(mdot_cup[i],3),"|",y_cup[i],"|"\
         ,round(vj_cup[i],3),"|",round(vd_cup[i],3),"|",round(mom_cup[i],3),"|",round(f_cup[i],3),"|")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fit a linear model: F = m*x + b
m, b = np.polyfit(mom_plate, f_plate, 1)
fit_line = m * np.array(mom_plate) + b

# Scatter plot of original data
plt.scatter(mom_plate, f_plate, label="$F(J)$ for the flat plate")

# Plot fitted line
plt.plot(mom_plate, fit_line,
         label=f"Fit: $F(J) = {m:.2f}J + {b:.2f}$",
         linestyle='-',color="red")

# Labels, title, legend, and grid
plt.xlabel("Momentum $J=\\dot{m}\\cdot u_0$ \n ${kgÂ·m^2}/{s^2}$")
plt.ylabel("Force $F=\\frac{M\\cdot g \\cdot (a+y)}{a}$\n $(kg~m/s^2)$")
plt.title("Flat Plate Vane")
plt.legend()
plt.grid(True)
plt.savefig("plate_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fit a linear model: F = m*x + b
m, b = np.polyfit(mom_cup, f_cup, 1)
fit_line = m * np.array(mom_cup) + b

# Scatter plot of original data
plt.scatter(mom_cup, f_cup, label="$F(J)$ for the hemispherical cup")

# Plot fitted line
plt.plot(mom_cup, fit_line,
         label=f"Fit: $F(J) = {m:.2f}J + {b:.2f}$",
         linestyle='-',color="red")

# Labels, title, legend, and grid
plt.xlabel("Momentum $J=\\dot{m}\\cdot u_0$ \n ${kgÂ·m^2}/{s^2}$")
plt.ylabel("Force $F=\\frac{M\\cdot g \\cdot (a+y)}{a}$\n $(kg~m/s^2)$")
plt.title("Hemispherical Cup Vane")
plt.legend()
plt.grid(True)
plt.savefig("cup_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```
