# Lab 0 - Role of Engineering Experimentation

The primary role of engineering experimentation is to progress from theory (dreamed up in our biological computers) to validation (confirm our theory is useful, and how accurate a predictor it may be).  

## Learning Objectives

By the end of this session, students will be able to:

- Explain the scientific method in historical and engineering contexts
- Distinguish between physical laws, empirical models, and measured data
- Apply model-based reasoning to experimental design
- Use error analysis to evaluate the agreement between theory and experiment

## Part 1: The Scientific Method — Historical and Practical Application

In this laboratory course, we repeatedly apply the scientific method as a process for testing and validating models of fluid behavior. 

Each experiment will follow a workflow that mimics elements of the scientific method; the workflow summary is:

1. **Model Definition**  We begin by identifying the governing physical principles and equations that define the theoretical model (e.g., Bernoulli’s equation, momentum balance, or head loss formulas).  In the experiments we are ultimately going to test the utility of the model to explain some real phenomenon.  The definition will involve some derivation, sketching, and analysis like you have already done in CE 3305.
2. **Model Predictions** Using known parameters and appropriate assumptions, we calculate what the model predicts should occur under the given experimental conditions. This component is conducted **before** conducting any experiments or making any measurements.
3. **Measurement and Observation** Here is where we conduct the experiment, collecting data through direct measurement of key variables such as flow rate, pressure, or velocity.
4. **Comparison and Error Analysis** We analyze how well the measured values align with the predicted results, using plots and error metrics such as percent error or RMS error to quantify agreement.
5. **Interpretation and Evaluation** Finally, we reflect on the model’s performance. Where results diverge from theory, we consider the influence of experimental error, unmodeled effects, or invalid assumptions, and discuss whether the model should be accepted, rejected, or refined.

This workflow turns each experiment into a rigorous test of theory, helping you build both conceptual understanding and practical insight into fluid mechanics.

### A Brief History of the Scientific Method

[Wikipedia: History of the Scientific Method](https://en.wikipedia.org/wiki/History_of_scientific_method) presents a good discourse on the history of the "method;" notable players of that history include:

- **Francis Bacon (1620s)**: Proposed structured observation and inductive reasoning.  The Baconian method is the investigative method developed by Francis Bacon, one of the founders of modern science, and thus a first formulation of a modern scientific method. The method was put forward in Bacon's book Novum Organum (1620), or 'New Method', to replace the old methods put forward in Aristotle's Organon. It influenced the early modern rejection of medieval Aristotelianism. He explicitly addressed causality in his writings on the subject.
- **Galileo Galilei**: Used mathematical models to describe physical phenomena, and designed experiments to test them. Galileo Galilei played a pivotal role in shaping the modern scientific method by emphasizing experimentation, observation, and mathematical analysis. He developed a method that involved intuition, demonstration, and experimental verification to test hypotheses and build upon existing knowledge. Galileo's work, particularly in physics and astronomy, demonstrated the power of combining observation, experimentation, and mathematical reasoning to understand the natural world
- **Isaac Newton**: Developed universal laws that combined observation, theory, and prediction. Isaac Newton's scientific method was a refined approach to scientific inquiry that emphasized observation, experimentation, and the use of mathematical reasoning to develop and test theories. He used a combination of theoretical reasoning and empirical observation to draw inferences and develop his groundbreaking theories of motion and gravitation.

### Adaptation to Engineering

The engineering design process is a structured approach to solving problems, building upon the principles of the scientific method but with a focus on practical application and creating tangible solutions. The primary tool is the *model* which relates the excitation-response behavior of the world to the phenomenon of interest.  Designers can rely upon the model once it is demonstrated to be reliable and accurate - the point of engineering experimentation is usually to test the model (for applicability, reliability, and accuracy). 

- **Theories** (e.g., conservation laws) → produce **models**.
- **Models** → generate **predictions** given specific inputs.
- **Experiments** → gather data to test how well predictions match reality.

:::{admonition} Structured Inquiry vs. "Cook-and-Look": Two Paths Through the Laboratory

**Overview**

Not all experiments are created equal. In engineering education, two styles of experimentation commonly emerge:

- **Scientific-method-directed experiments**, where theory guides the inquiry, predictions are made, and results are critically evaluated to test a hypothesis or validate a model.
- **"Cook-and-look" experiments**, where a procedure is followed step-by-step with little theoretical framing or expectation, often leading to passive observation rather than active discovery.

Understanding the distinction is essential not only for academic success, but for cultivating the mindset of a practicing engineer.

---

**Scientific-Method-Directed Experimentation**

The characteristics of scientific-method directed inquiry are:

- Begins with a hypothesis or model derived from theory
- Predictions are made in advance
- Experiments are **designed** to test, calibrate, or refute the model
- Measurements are compared to expectations with quantitative rigor
- Discrepancies lead to refinement of models and deeper understanding

Some noteable successes attributed to structured inquiry are:

- **Galileo’s inclined plane experiments (1600s)**  Tested mathematical relationships between acceleration, time, and distance, providing early evidence for the laws of motion.
- **Cavendish’s measurement of the gravitational constant (1797)**  Designed to quantify Newton’s law of universal gravitation—not merely observe attraction.
- **Prandtl’s boundary layer experiments (early 1900s)**  Theoretical analysis led to the design of experiments that revealed the structure of viscous flow, revolutionizing fluid mechanics.

Notable failures include:

- **Michelson–Morley experiment (1887)**  Failed to detect the "aether wind" predicted by classical physics. Though it produced a “null result,” it motivated a shift in thinking that led to Einstein’s theory of relativity.  It is an example of a model built on faulty assumptions, hence the experiment could not validate the model, researchers modified assumptions over time until Einstein's model(s) indeed could explain a particular behavior.

---

**"Cook-and-Look" Experimentation**

The characteristics of "cook-and-look" directed inquiry are:

- Follows a fixed set of steps like a recipe, often from a lab manual
- **No** hypothesis or model-driven **prediction** is made
- Data is collected without clear expectations
- Emphasis is on procedure compliance, not interpretation
- Often ends with, “The results didn’t match, but we followed the steps.”

Some occasional successes attributed to cook-and-look are:

- **Alexander Fleming’s discovery of penicillin (1928)** Accidental contamination of a petri dish revealed antibacterial effects—not predicted by theory, but recognized through observation.

- **Discovery of cosmic microwave background (Penzias and Wilson, 1965)**  Researchers were troubleshooting antenna noise and stumbled onto the remnant of the Big Bang—without initially understanding the implications.

These examples show that "cook-and-look" can occasionally lead to breakthroughs, but only if the observer is prepared to recognize the unexpected and reframe the context.

More common are failures, where the results are of little value:

- Countless teaching, research, and industrial laboratories where results are recorded without interpretation.
- Industrial settings where procedures are followed, but underlying phenomena are misunderstood, leading to poor troubleshooting or design decisions

> Bear in mind that cook-and-look experiments are totally appropriate for certain compliance purposes, in such instances the rote application of process is important to determine if something is to be certified to a standard or not.  Arguably there is scientific method involved, because the standards to be met are the "model", and if the experiment fails, the component is inadequate.

>This type of experimentation emphasizes verification over discovery, and the value lies in reproducibility and adherence to established thresholds rather than hypothesis generation. While it may lack the iterative curiosity of traditional scientific inquiry, it still involves rigorous observation, data collection, and comparison to a known benchmark. In this context, the "model" is codified in standards documents—such as ASTM or ISO—and the experiment's purpose is to determine conformance, not to explore new phenomena.

## Why Scientific Method Matters in Engineering

Engineers do not merely measure—they model, predict, and verify. Experiments are not recipes; they are **tools to challenge assumptions**, test applicability, and build confidence in our understanding of physical systems.

In this laboratory course, we reject the "cook-and-look" mindset. Instead, we adopt a **model-driven approach** that values prediction, observation, comparison, and critical thinking. Your role is not to confirm the manual—it is to test nature and decide whether our models are worthy of your trust.

# Course Purpose
Engineering relies on models to represent how the physical world behaves under various conditions. These models may be derived from physical laws (e.g., conservation of mass or energy), empirical observations, or simplifications of complex phenomena.

The primary purpose of this fluids laboratory course is not simply to repeat known formulas or perform routine measurements—it is to **test, calibrate, and validate models of fluid behavior** through structured experimentation.

:::{admonition} Why a Laboratory Experience?

A laboratory experience in engineering education provides students with hands-on opportunities to apply theoretical knowledge, develop practical skills, and foster critical thinking and problem-solving abilities essential for professional and academic success.

:::

In this course, we will follow a **consistent** experimental process rooted in the scientific method:

1. **Develop/Review the governing theory.**  Identify the underlying model for the system under investigation—whether it's Bernoulli's equation, the Darcy-Weisbach equation, or a momentum balance.

2. **Predict behavior based on the model.**  Before touching any equipment, calculate what should happen based on the inputs and assumptions. This includes identifying dependent variables, known parameters, and expected relationships.

3. **Measure the system’s behavior experimentally.**  Use lab equipment to collect data. Take care to understand how instruments work, and estimate uncertainty where appropriate.

4. **Compare model predictions with measurements.**  Plot your results, compute error metrics (such as percent error, root-mean-square error, or residuals), and assess whether the model agrees with reality.

5. **Evaluate and reflect.**  Consider whether deviations are due to limitations in the model, instrumentation error, procedural issues, or real-world complexity. Can the model be recalibrated? Was a core assumption invalid?
