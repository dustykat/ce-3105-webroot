## Additional Guidance

Submission Guidelines for Lab Reports

The instructor uses a Language Learning Model (OnPrem.LLM; Modular, privacy-first, local document & prompt handling) to generate feedback and provide an initial score for your work. To ensure accurate evaluation and maximize your score:

### File Naming Convention
Follow the explicit file name convention provided by the instructor (e.g., LastName_FirstName_AssignmentName.pdf). This ensures your work is identifiable.

### File Format
Submit your work as a PDF file. If your assignment involves computational scripts or code, include them as clearly formatted appendices in the same PDF or submit a zipped folder containing both the PDF and the scripts.

### Structure Your Document 

Use the report format provided.

### Use Consistent Formatting

- Use a standard font (e.g., Times New Roman, 12pt) and 1.5x line spacing.
- Use headings for major sections (e.g., "Problem 1," "Solution").
- Label and reference all figures, tables, and equations.

### Code Integration

- If applicable, embed short code snippets directly into the document and ensure they are legible.
- For longer scripts, attach them in an appendix or as separate files.

### Calculations and Units

- Show all steps in your calculations (representative calculations are preferable -  I assume you will use a script or spreadsheet for repetitive calculations).
- Use consistent units throughout and verify that units are included in all results.

### Graphs and Diagrams

- Clearly label all axes and include titles. [**Graphics Conventions (link)**](http://54.243.252.9/engr-1330-webroot/engr1330jb/_build/html/lessons/lesson14/lesson14.html)
- Add legends if multiple datasets are plotted.
- Ensure diagrams are annotated and relevant to your solution.

### References

Cite all external materials used (e.g., textbooks, research papers, or online resources).
