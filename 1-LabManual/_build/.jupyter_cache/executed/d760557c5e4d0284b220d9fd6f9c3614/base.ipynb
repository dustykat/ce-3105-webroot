```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
V   = np.array([0.0218, 0.0321, 0.0546, 0.0742, 0.0784])
DH  = np.array([0.035, 0.057, 0.103, 0.201, 0.199])

# Create plot
fig, ax1 = plt.subplots(figsize=(8, 5))

# Primary axis: FF1 vs RE
ax1.plot(V, DH, 'o' ,color='tab:blue', label='Head Loss')
ax1.set_xlabel('Velocity (m/s)')
ax1.set_ylabel('Head Loss (m)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.grid(True)

plt.title('Head Loss vs Velocity')
plt.tight_layout()
plt.show()
```

```python
# Compute Area for a Circular Pipe
def area_circle(diameter):
    import math
    area_circle = (math.pi*diameter**2)/4.0
    return area_circle
```

```python
print(f"Pipe Flow Area: {area_circle(0.003):10.3e} square meters")
```

```python
# Compute Reynolds Number
def reynolds(velocity,diameter,kinematic):
    reynolds = velocity*diameter/kinematic
    return(reynolds)
```

```python
print(f"Reynolds Number: {reynolds(0.0218,0.003,1.14e-06):10.1f} ")
```

```python
# Compute Friction Factor given Headloss
def fricfact(length,diameter,velocity,gravity,headloss):
    fricfact = headloss*(diameter/length)*((2.0*gravity)/velocity**2)
    return(fricfact)
```

```python
print(f"Friction Factor: {fricfact(0.524,0.003,0.0218,9.81,0.035):10.2f} ")
```
