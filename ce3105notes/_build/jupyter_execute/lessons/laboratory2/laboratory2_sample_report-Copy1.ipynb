## **Introduction**
The purpose of this laboratory experiment was to investigate fluid statics through two key objectives:
1. Measure the buoyancy force on various objects and verify Archimedes’ Principle.
2. Determine the hydrostatic thrust acting on a plane surface using a quadrant balance, for both partially and fully submerged conditions.

## **Materials and Methods**

### **Materials**
- Quadrant balance apparatus
- Graduated cylinder
- Thermometer
- Objects for buoyancy testing (rocks, composites, wood samples)
- Weighing scale
- Water
- Transfer pipette
- Standard weights and hangers

### **Methods**
#### Part 1: Buoyancy Experiments
1. Measured the temperature of water and recorded initial volumes in a graduated cylinder.
2. Submerged objects (Rock-1, Rock-2, etc.) and recorded displaced volumes.
3. Repeated measurements three times for each object to ensure accuracy.
4. Recorded data for each object, including mass and displacement.

#### Part 2: Hydrostatic Forces
1. Set up the quadrant balance and trimmed it to ensure the plane surface was vertical.
2. Conducted trials for partially submerged conditions by adjusting weights and water levels.
3. Conducted trials for fully submerged conditions by further submerging the plane surface.
4. Recorded water depths $ h $ and widths $ b $ of the free surface for all trials.

## **Data Analysis**

### **Part 1: Buoyancy**

The process began by calculating the displaced volume, $ \Delta V $, for each object, as detailed in the tables above. 
Using these volumes, buoyancy forces were determined through the equation

$$ F_B = \rho_{\text{water}} \cdot \Delta V \cdot g $$

which provided a basis for comparing the calculated object volumes with those derived from geometric measurements. 

The agreement between these methods was comparable, particularly for objects with spherical or prismatic geometries. 

Finally, Archimedes’ Principle was verified for both floating and submerged objects by confirming that, for submerged items, the weight force exceeded the computed buoyant force.


### **Part 2: Hydrostatic Forces**

The process began by calculating moments $ M $ for all trials using:
   
   $$ M = W \cdot \left( \frac{3b}{8} \right) \cdot h $$

Then a plot $ M $ vs. $ h $ for fully submerged trials was created (below) and the slope was determined from a trendline. 

```python
import matplotlib.pyplot as plt
import numpy as np
# Given data
_M = [0.501, 0.450, 0.424, 0.379, 0.334]
_h = [78, 70, 66, 59, 52]
# Compute h^3 for each value in _h
_h_cubed = np.array(_h) ** 3
# Scatter Plot with Trendline
# Fit a linear model (trendline) to the data: M = m*(h^3) + b
coeffs = np.polyfit(_h, _M, 1)  # Linear fit (degree=1)
trendline = np.poly1d(coeffs)
# Generate a set of x-values for the trendline
x_vals = np.linspace(min(_h), max(_h), 100)
y_vals = trendline(x_vals)
#
slope = float(trendline[(1)])
slope = format(slope, ".3e")
intercept = float(trendline[(0)])
#
plottitle = 'Fully Submerged M vs $h$\n' + 'Slope = ' + slope + '\n Intercept = ' + repr(round(intercept,3))
plt.figure(figsize=(8, 4))
plt.plot(_h, _M, '.', label='Data points')
plt.plot(x_vals, y_vals, '-', label='Trendline')
plt.xlabel('$h$')
plt.ylabel('M')
plt.title(plottitle)
plt.legend()
plt.grid(True)
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

3. Evaluated partially submerged data by plotting:

```python
import matplotlib.pyplot as plt
import numpy as np
# Given data
_M = [0.742, 0.752, 0.737, 0.735, 0.735]
_hh = [138, 135, 129, 125, 120]
# Compute h^3 for each value in _h
_h_cubed = np.array(_hh) ** 3
# Scatter Plot with Trendline
# Fit a linear model (trendline) to the data: M = m*(h^3) + b
coeffs = np.polyfit(_h_cubed, _M, 1)  # Linear fit (degree=1)
trendline = np.poly1d(coeffs)
# Generate a set of x-values for the trendline
x_vals = np.linspace(min(_h_cubed), max(_h_cubed), 100)
y_vals = trendline(x_vals)
#
slope = float(trendline[(1)])
slope = format(slope, ".3e")
intercept = float(trendline[(0)])
#
plottitle = 'Partial Submerged M vs $h^3$\n' + 'Slope = ' + slope + '\n Intercept = ' + repr(round(intercept,3))
plt.figure(figsize=(8, 4))
plt.plot(_h_cubed, _M, '.', label='Data points')
plt.plot(x_vals, y_vals, '-', label='Trendline')
plt.xlabel('$h^3$')
plt.ylabel('M')
plt.title(plottitle)
plt.legend()
plt.grid(True)
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

## **Discussion**
The experimental results confirmed Archimedes’ Principle: the buoyant forces **measured** aligned well with the theoretical predictions based on displaced water volumes, with floating objects adhering closely to the calculated masses. Additionally, the analysis of hydrostatic forces revealed that the experimental measurements of the center of pressure and the resulting moments were largely consistent with theoretical expectations, with only minor deviations likely caused by the precision of the setup and measurement errors. The study also highlighted potential improvements, suggesting that better calibration of the quadrant balance and an increased number of trials could further enhance the accuracy of future results.
