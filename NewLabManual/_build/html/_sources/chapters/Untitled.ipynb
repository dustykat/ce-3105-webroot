Graphics Conventions

Clear and accurate graphics are essential in engineering communication. In professional practice, figures are not decoration‚Äîthey are part of the analytical record. A poorly labeled graph or ambiguous diagram can lead to misinterpretation of data, incorrect decisions, or regulatory delays. The conventions below mirror standards used in consulting reports, design memoranda, and technical documentation. Follow them exactly to ensure your results can be understood, checked, and used by others.

1. Axes and Units

Every axis must include both a variable name and units (e.g., Velocity, m/s).

Units must follow SI conventions unless otherwise stated.

Use consistent scaling‚Äîavoid distortions that exaggerate trends.

2. Figure Titles and Captions

Every figure requires a figure number and descriptive title.

Example:
Figure 3. Force‚ÄìMomentum Relationship for the Flat Plate Vane

Captions should explain:

What is being shown

Why it is significant

How to interpret key features

Which equation(s) the plot supports or tests

3. Readability Standards

Minimum font size: 10‚Äì11 pt for axis labels, tick labels, and legends.

Line thickness ‚â• 1.0 pt.

Use color sparingly; ensure contrast is readable when printed in grayscale.

4. Legends and Multiple Datasets

Include legends whenever more than one dataset is shown.

Ensure symbols, colors, or line-styles are distinct.

5. Data Integrity

Do not manipulate or remove points unless justified.

Outliers must be discussed, not simply deleted.

Trendlines must use a defensible model (e.g., linear regression for linear theory).

6. Referencing Figures in Text

Every figure must be referenced at least once in the report.

Example:
‚ÄúAs shown in Figure 3, the flat plate exhibits a nearly linear relationship between momentum flux and force.‚Äù

7. Diagram Requirements

Schematic diagrams should show:

Relevant geometry

Measurement locations

Boundary conditions (if applicable)

Key dimensions labeled

Use straight lines, clean arrows, and uniform fonts.

8. LLM-Friendly Practices

Enclose figure captions in standard formats so the scoring engine can detect them.

Avoid pasted screenshots of tables‚Äîuse real text-based tables.

Ensure numerical data underlying your graph comes from machine-readable cells or scripts.

Graphics Standards for CE 3105 Reports

In professional hydraulics and fluid mechanics work, figures and diagrams frequently drive design decisions, safety evaluations, and regulatory approvals. Poorly constructed graphics create ambiguity, slow down reviews, and undermine engineering credibility. To reflect industry expectations, all figures in CE 3105 must adhere to the following standards:

A. Technical Accuracy

Figures must accurately reflect the data collected.

No unstated filtering, averaging, or omission of data points.

All regression lines must be mathematically justified and labeled.

B. Communication Standards

All axes: variable + units (SI only).

Number and title every figure in order of appearance.

Include succinct captions that aid interpretation.

Reference each figure by number in the main body of the report.

C. Style Standards

Use consistent font and color choices across all figures.

Maintain readability when printed in grayscale.

Keep diagrams simple and functional‚Äîavoid unnecessary graphics.

D. Reproducibility & Transparency

Underlying data used for plots must be present in the Results or Appendix.

Data transformations must be described in the text or caption.

Code used to produce plots must be included in the Appendix.

E. Professional Integrity

All figures must represent your team‚Äôs data.

External images require citation and must not substitute for required measurements.

Plagiarism in figures (including copied schematics) is treated the same as text plagiarism.

These standards are aligned with expectations you will encounter in design offices, consulting engineering reports, hydrologic/hydraulic modeling documentation, and formal submissions to state or federal agencies.
