# Laboratory Reports

In engineering practice, written reports are the primary means by which technical work is communicated, reviewed, and relied upon. Decisions involving design, operations, safety, and compliance depend on the engineer’s ability to document methods, present data clearly, justify conclusions, and explain the reasoning behind them. Laboratory reports serve as early, structured training in these professional expectations. They ensure that the work you perform in the lab can be understood, reproduced, and critically evaluated by others—just as it must be in real engineering environments.

The purpose of this report format is to develop the habits of communication that engineers use throughout their careers. A strong report demonstrates more than completed calculations: it demonstrates understanding of the underlying principles, thoughtful analysis of results, accountability in data handling, and clarity in presenting technical information. The sections below outline the standard structure for laboratory reporting in this course. Each component mirrors a function performed in industry—from stating an objective to supporting conclusions with evidence—so that your work reflects both academic rigor and professional practice.

## Report Requirements
Each student must submit a lab report consisting of the following items:
1. Title Page
   - Include the course number, section number, and course title, descriptive title of the experiment, date completed, date submitted, and prepared by (author‟s
name) and submitted to (lab instructor‟s name)
2. Table of Contents
   - List major report headings;
   - Show a list of tables and figures with titles;
   - Indicate the contents of the appendix ;
   - Show corresponding page numbers 
3. Objective
   - Clearly state the purpose of the investigation
4. Theory
   - Discuss fundamental principles behind the experiment;
   - Clarify equations to be used and define all variables;
   - Indicate important assumptions for validity of equations;
   - Indicate your anticipated results;
   - Cite references for borrowed ideas or expressions
5. Apparatus
   - Drawings of the apparatus with important dimensions
6. Procedure
   - A concise list of the steps your group performed
7. Results
    - All your results should be presented here;
    - Tables, charts and plots should be incorporated into the report text;
    - For each table or figure, provide a brief narrative that explains the significance of the table or figure, its content, and, by referencing equations in the “Theory” section, how values were obtained;
    - Provide a number and title for each table and figure that corresponds to that given above
8. Discussion
    - Discuss the accuracy and precision of each result;
    - Analyze probable sources of error
9. Conclusion
    - State and justify whether your initial objective was met;
    - Can you make recommendations for future related work
10. References
    - List of references cited in the report
11. Appendix
    - Original, handwritten data tables (if appropriate);
    - Example hand calculations;
    - Scripts (python code used to analyze data and produce plots)

## Submision Guidelines

The instructor uses a Language Learning Model (OnPrem.LLM; OpenAI GPT 5+; or similar tools) to screen your reports, generate actionable feedback and provide an initial score for your work. To ensure meaningful evaluation and to maximize your score:

1. Follow the explicit file name convention below:

:::{admonition} File Naming Convention
:class: student-live

The file you upload should be named: **LastName_FirstName_AssignmentName.pdf**

This ensures your work is 
1. Identifiable for scoring 
2. Lexiographically sortable 
:::

2. Use the correct file format:

:::{admonition} File Format
:class: student-live

The file you upload should be formatted as a PDF file. If your assignment involves computational scripts or code, include them as clearly formatted appendices in the same PDF.  Data tables should be machine readable (not images or screen captures).  Figures must be captioned so the LLM can detect their presence, and attept to interpret them (its hard for LLM to interpret images, but if identified as a Figure there is a better chance).  

Hand-written content is nearly impossible to interpret, so type your reports.
:::

3. Structure Your Document 

   - Use the report format provided - verbatim.  Even if some components are empty, include the headings in your report.

4. Use Consistent Formatting
   - Use a standard font (e.g., Times New Roman, 12pt) and 1.5x line spacing.
   - Use headings for major sections (e.g., "Problem 1," "Solution").
   - Label and reference all figures, tables, and equations.

5. Code Integration
   - If applicable, embed short code snippets directly into the document and ensure they are legible. If you use Word to build your report use Courier font for code fragments.
   - For longer scripts, attach them in the appendix .

6. Calculations and Units
   - Show all steps in your calculations (representative calculations are preferable -  We assume you will use a script or spreadsheet for repetitive calculations).
   - Use consistent units throughout and verify that units are included in all results.

7. Graphs and Diagrams
   - Clearly label all axes and include titles. [**Graphics Conventions (link)**](http://54.243.252.9/engr-1330-webroot/engr1330jb/_build/html/lessons/lesson14/lesson14.html)
   - Add legends if multiple datasets are plotted.
   - Ensure diagrams are annotated and relevant to your solution.

8. References
   - Cite all external materials used (e.g., textbooks, research papers, or online resources).

:::{admonition} Why this matters
Engineering is a profession built on clarity, accountability, and the ability to communicate technical work in a way that others can trust and act upon. A laboratory report is more than a record of what you did in the lab—it is a demonstration of the same skills you will rely on in professional practice: organizing information, documenting methods, presenting data accurately, justifying decisions, and explaining results in a way that another engineer can understand, reproduce, and evaluate.

The structure and expectations outlined in this template are not arbitrary; they mirror the documentation standards used in design offices, consulting firms, research organizations, and regulatory environments. Developing these habits now prepares you for work where reports influence real decisions about safety, performance, costs, and compliance. Writing a clear, complete, and well-reasoned report is therefore not just a course requirement—it is foundational to becoming a competent and trustworthy engineer.
